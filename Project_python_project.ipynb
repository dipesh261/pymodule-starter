```python
# 1. How many unique customer IDs are present in the dataset?
unique_customers = df['Customer ID'].nunique()
print("Unique Customers:", unique_customers)
```

```python
# 2.What is the total sales amount in the dataset?
total_sales = df['Sales'].sum()
print("Total Sales:", total_sales)
```

```python
#3.Average sales amount per order
average_sales_per_order = df.groupby('Order ID')['Sales'].sum().mean()
print("Average Sales per Order:", average_sales_per_order)
```

```python
# 4. Orders placed in each region
orders_per_region = df.groupby('Region')['Order ID'].nunique()
print("Orders per Region:")
print(orders_per_region)
```

```python
#5.Distribution of orders by ship mode
order_distribution_by_ship_mode = df['Ship Mode'].value_counts()
print("Order Distribution by Ship Mode:")
print(order_distribution_by_ship_mode)
```

```python
# 6. City with the maximum number of orders
city_with_max_orders = df['City'].value_counts().idxmax()
print("City with the Maximum Number of Orders:", city_with_max_orders)
```

```python
# 7. State with the highest sales revenue
state_with_highest_sales = df.groupby('State')['Sales'].sum().idxmax()
print("State with the Highest Sales Revenue:", state_with_highest_sales)
```

```python
# 8. Total products sold in the "Furniture" category
furniture_products_sold = df[df['Category'] == 'Furniture']['Product ID'].nunique()
print("Total Products Sold in Furniture Category:", furniture_products_sold)
```

```python
# 9. Percentage of total sales from the "Corporate" segment
total_sales_corporate = df[df['Segment'] == 'Corporate']['Sales'].sum()
total_sales_percentage_corporate = (total_sales_corporate / df['Sales'].sum()) * 100
```

```python
total_sales_percentage_corporate
```

```python
#10. Orders shipped on the same day as the order date
orders_shipped_same_day = df[df['Ship Date'] == df['Order Date']]['Order ID'].nunique()
print("Orders Shipped on the Same Day as Order Date:", orders_shipped_same_day)
```

```python
#11. Unique products listed
unique_products = df['Product ID'].nunique()
print("Unique Products Listed:", unique_products)
```

```python
#12. Most frequently purchased sub-category
most_frequent_sub_category = df['Sub-Category'].value_counts().idxmax()
print("Most Frequently Purchased Sub-Category:", most_frequent_sub_category)
```

```python
#13. Customer with the highest number of orders
customer_with_highest_orders = df['Customer ID'].value_counts().idxmax()
print("Customer with the Highest Number of Orders:", customer_with_highest_orders)
```

```python
#14. Average sales amount per category
average_sales_per_category = df.groupby('Category')['Sales'].mean()
print("Average Sales Amount per Category:")
print(average_sales_per_category)
```

```python
#15. Mlesedian sales value for the "Office Supplies" category
median_sales_office_supplies = df[df['Category'] == 'Office Supplies']['Sales'].median()
print("Median Sales Value for Office Supplies Category:", median_sales_office_supplies)
```

```python
#16. Sales distribution across regions
sales_distribution_by_region = df.groupby('Region')['Sales'].sum()
print("Sales Distribution by Region:")
print(sales_distribution_by_region)
```

```python
#17. Top 5 products by sales
top_5_products_by_sales = df.groupby('Product Name')['Sales'].sum().nlargest(5)
print("Top 5 Products by Sales:")
print(top_5_products_by_sales)
```

```python
#18. Total sales for each shipping mode
total_sales_by_shipping_mode = df.groupby('Ship Mode')['Sales'].sum()
print("Total Sales by Shipping Mode:")
print(total_sales_by_shipping_mode)
```

```python
#19. Top 3 customers with the highest total sales
top_3_customers_by_sales = df.groupby('Customer ID')['Sales'].sum().nlargest(3)
print("Top 3 Customers by Sales:")
print(top_3_customers_by_sales)
```

```python
#20. Region with the least total sales
region_with_least_sales = df.groupby('Region')['Sales'].sum().idxmin()
print("Region with the Least Total Sales:", region_with_least_sales)
```
